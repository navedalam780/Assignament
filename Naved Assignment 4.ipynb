```python
## 1. List and demonstrate the use of the following arithmetic operators in Python: Addition, Subtraction, Multiplication, Division, Modulus, Exponentiation, Floor Division.

# Addition
result= 5+3
print(result)

# Subtraction
result= 5-3
print(result)

# Multiplication
result= 5*3
print(result)

# Division
result= 6/3
print(int(result))

# Modulus
result= 10%3
print(result)

# Exponentiation
result= 2**3
print(result)

# Floor Division (Gives the result value without decimal)
result= 20//3
print(result)
```

```python
## 2. Write Python expressions using comparison operators (less than, greater than, equal to, not equal to) and evaluate them. Include examples with both numbers and strings.

# (i)with numbers

# Less Than
print(5<2)

# Greater Than 
print(5>2)

# Equal To
print(5==2)

# Not Equal To 
print(5!=2)


# (i)with strings

# Less Than
print("apple" < "banana")  # Output: True (kyunki "apple" alphabetically "banana" se pehle aata hai)

# Greater Than 
print("grape" > "apple")  # Output: True (kyunki "grape" alphabetically "apple" ke baad aata hai)

# Equal To
print("orange" == "orange")  # Output: True

# Not Equal To 
print("apple" != "banana")  # Output: True
```

```python
# 6. Create a string variable and demonstrate the following string methods: slicing, negative indexing, converting to uppercase, converting to lowercase, stripping whitespace, replacing a substring, and splitting the string.

test_string = "Mera naam Naved Alam hai"

for_striping = "             Naved Alam           "

for_spliting = "Naved Alam"

# Slicing 
print(test_string[10:15])

# Negetive Indexing 
print(test_string[-14:-4])

# Converting to Upper Case
print(test_string.upper())

# Converting to Lower Case
print(test_string.lower())

# Striping Whitespace
print(for_striping.strip())

# Replace a Substring
print(test_string.replace("Naved Alam", "Sufiyan Khan"))

# Split The String
print(for_spliting.split(" "))   # Here is a space for spliting Which splits my name as first "Naved" and second "Alam". 
```

```python
# 9. Modify elements of a list by using the append(), insert(), remove(), pop(), and del methods. Print the list after each operation.

my_list= [10,20,30,50,60]

# Append a Number & Print
my_list.append(70)
print(my_list)

# Insert a Number & Print
my_list.insert(3,40)
print(my_list)

# Remove a Number & Print
my_list.remove(50)
print(my_list)

# Pop a Number & Print (Pop Function Value Remove karne ke saath saath remove ki huyi value return bhi karta hai aur isme element delete karne ke liye index number diya jaata hai)

poped_value = my_list.pop() #isme agar koi argument nahin diya to ye aakhri value ko uda deta hai 
print(poped_value)
print(my_list)


poped_value = my_list.pop(1) #isme agar koi index diya hai to wahi index wali value uda dega, Example ke liye my_list me se mujhe 20 wala element udana hai to aise likhenge 
print(poped_value)
print(my_list)
```
